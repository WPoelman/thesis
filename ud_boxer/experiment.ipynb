```python
def flatten_list(nested_list):
    """
    Flatten the given nested list if there are nested lists, otherwise return the list.
    """
    flattened_list = []
    for item in nested_list:
        if isinstance(item, list):
            flattened_list.extend(flatten_list(item))
        else:
            flattened_list.append(item)
    return flattened_list

flatten_list(['a', [2,3,3], 'b'])
```

```python
import re
import penman
import re

def remove_unwanted_quotes(s):
    # Replace spaces within quotes with underscores
    s = re.sub(r'"(.*?)"', lambda m: '"' + m.group(1).replace(' ', '_') + '"', s)
    
    # Remove unwanted quotes
    result = []
    i = 0
    while i < len(s):
        c = s[i]
        if c == '\\' and i + 1 < len(s) and s[i + 1] == '"':
            result.append('"')
            i += 2
        elif c == '"' and (i == 0 or s[i-1] != '\\'):
            i += 1
        else:
            result.append(c)
            i += 1
    return ''.join(result)


def replace_colon_in_time(match):
    return match.group(0).replace(':', '-')

def error_analysis(output):
    with open(output, 'r') as pred:
        lines = pred.readlines()
        target = []
        discourses =[]
        for i, line in enumerate(lines):
            if line != '\n':
                line = line.strip()
                line = line.replace("<root>", "")
                line = re.sub(r'\b\d{2}:\d{2}\b', replace_colon_in_time, line)
                new_line = remove_unwanted_quotes(line)
                new_list = [x for x in new_line.split(' ') if x in NEW_BOX_INDICATORS]
                if new_list:
                    discourses.append((i, new_line))       
                real_penman = penman.encode(penman.decode(new_line))
                target.append(real_penman)
    return target, discourses

preds, pred_discourses = error_analysis('parserOut.txt')
refs, refs_discourses = error_analysis('gold_dev.txt')
```
